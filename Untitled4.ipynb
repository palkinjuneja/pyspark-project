```python
###### IMPORTING LIBRARIES ##################
import findspark
findspark.init('/home/palkin/spark-2.4.0-bin-hadoop2.7')
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('telecom').getOrCreate()
import matplotlib.pyplot as plt
import pandas
from pyspark.sql.functions import col
%matplotlib qt
```

```python
####### creting pyspark dataframe ###########
new_df=spark.read.csv("/home/palkin/Downloads/call_data.csv",inferSchema=True,header=True)#### call
new_df2=spark.read.csv("/home/palkin/Downloads/speed_data.csv",inferSchema=True,header=True)### speed 
p=new_df.count()
q=new_df2.count()

print ("total rows in voice data is",p)
print ("total rows in speed data is",q)
###### drop null values
dna_df=new_df.na.drop()
dna_df1=new_df2.na.drop()
#dna_df2=new_df2.drop
dna_df2=dna_df1.filter(dna_df1['LSA'].isin(['NA'])=='FALSE' )
dna_df2.count
#p.show()
#dna_df2.show()
```

```python
from pyspark.sql.functions import col, desc

def user_vs_provider(df,t,title):
    
    len_call=df.count()
    
    total_call=df.groupBy('Operator').count()
    total_call_ratio = total_call.withColumn('count',total_call['count']/len_call)
    total_call_ratio.show()
    p_total_call=total_call_ratio.toPandas()
    p_total_call.plot.bar('Operator','count')
    plt.title(title)
    plt.xlabel('OPERATORS')
    plt.ylabel('Ratio Of Users')
    plt.show()
    top_op=total_call_ratio.sort(col("count").desc()).limit(5)
    pd=top_op.toPandas()
    return pd
# QUES1: Which operator is leading the market in terms of call?
opc=user_vs_provider(dna_df,'call','Operator wise distribution of total user (for voice quality data)')
print(opc)

# QUES1: Which operator is leading the market in terms of speed?
ops=user_vs_provider(dna_df2,'speed','Operator wise distribution of total user (for speed data)')
print(ops)
```

```python
def user_vs_state(df,t,title):
    if(t=='call'):
        s_name='State Name'
    elif(t=='speed'):
        s_name='LSA'
    
    len_call=df.count()
    total_call=df.groupBy(s_name).count()
    total_call_ratio = total_call.withColumn('count',total_call['count']/len_call)
    total_call_ratio.show()
    p_total_call=total_call_ratio.toPandas()
    p_total_call.plot.bar(s_name,'count')
    plt.xlabel(' STATE NAME')
    plt.ylabel('Ratio Of Users')
    plt.title(title)
    plt.show()
    top_op=total_call_ratio.sort(col("count").desc()).limit(5)
    v=top_op.toPandas()
    print(v)
    return v
# QUES1: Which operator is leading the market in terms of call?
stc=user_vs_state(dna_df,'call','state wise distribution of total users ( for voice quality data)')
# QUES1: Which operator is leading the market in terms of speed?
sts=user_vs_state(dna_df2,'speed','state wise distribution of total users ( for speed data)')
```

```python
def user_vs_ntype(df,t,title):
    if(t=='call'):
        s_name='Network Type'
    elif(t=='speed'):
        s_name='technology'
    df2=df.filter(df[s_name].isin(['Unknown'])=='FALSE')
    len_call=df2.count()
    total_call=df2.groupBy(s_name).count()
    total_call_ratio = total_call.withColumn('count',total_call['count']/len_call)
    total_call_ratio.show()
    p_total_call=total_call_ratio.toPandas()
    p_total_call.plot.bar(s_name,'count')
    plt.xlabel('NETWORK TYPE')
    plt.ylabel('Ratio Of Users')
    plt.title(title)
    plt.show()

# QUES1: Which operator is leading the market in terms of call?
user_vs_ntype(dna_df,'call',' Network type vs Users(for voice quality)')
# QUES1: Which operator is leading the market in terms of speed?
user_vs_ntype(dna_df2,'speed','Network Type vs Users (for speed data)')
```
